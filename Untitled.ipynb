```python
import requests
rusd = requests.get("https://bank.gov.ua/NBU_Exchange/exchange_site?start=20210101&end=20211231&valcode=usd&sort=exchangedate&order=desc&json")
reur = requests.get("https://bank.gov.ua/NBU_Exchange/exchange_site?start=20210101&end=20211231&valcode=eur&sort=exchangedate&order=desc&json")

reur.json()
```

```python
import json

usd = rusd.json()
eur = reur.json()
print(len(eur))
```

```python

with open("HEREUR.json", "w") as outfile:
    json.dump(eur, outfile)
    print(outfile)
outfile.close()
with open("HERUSD.json", "w") as outfile:
    json.dump(usd, outfile)
    print(outfile)
```

```python

import csv
 
 
data_file = open('jsonoutput.csv', 'w', newline='')
csv_writer = csv.writer(data_file)


 
count = 0
for i in range(0, len(reur.json()), 1):
    datae = eur[i]
    datau = usd[i]
    if count == 0:
        header = datae.keys()
        csv_writer.writerow(header)
        count += 1
    csv_writer.writerow(datae.values())
    csv_writer.writerow(datau.values())
    
```

```python
import matplotlib.pyplot as plt
import csv

monthlen=[31,28,31,30,31,30,31,31,30,31,30,31]

def datetrans(date):
    res=int(date[0]+date[1])
    for i in range(0, int(date[3]+date[4])-1, 1):
        res+=monthlen[i]
    return res
  
x = []
yusd = []
yeur = []
  
with open('res/jsonoutput.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    counter=0
    for row in lines:
        if counter == 0: 
            counter+=1
            continue
        if row[2]=='EUR':
            yeur.append(float(row[5]))
            x.append(datetrans(row[0]))
        else: yusd.append(float(row[5]))
  
plt.plot(x, yeur, color = 'orange', label='EUR')
plt.plot(x, yusd, color = 'g', label='USD')
plt.xlabel('Dates')
plt.ylabel('Currency price')
plt.title('Exchange Rates', fontsize = 20)
plt.show()
```
